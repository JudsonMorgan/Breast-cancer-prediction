Features with high correlation means that both features has the same effect on the dependent variable. In this case, we can likely drop one of the features.

```python
columns = ['area_mean', 'concavity_mean', 'concave points_mean', 'radius_worst', 'perimeter_worst',
'area_worst', 'concave points_worst', 'fractal_dimension_worst', 'compactness_worst', 'concave points_se',
'fractal_dimension_mean', 'Target' ]

df_update = df[columns]

df_update.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
df_update['Target'].value_counts()
```

```python
X = df_update.iloc[:, : -1]
y = df_update.iloc[:, -1]
```

```python
counter = Counter(y)
print(counter)
```

```python
sm = SMOTE()
X, y = sm.fit_resample(X, y)
counter = Counter(y)
print(counter)
```

```python
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
X_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state = 42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
reg = LogisticRegression()
reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
reg.score(X_train, y_train)
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_test, y_test)
```

```python
rfc.score(X_train, y_train)
```

### Evaluation

```python
from sklearn.metrics import accuracy_score, balanced_accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
y_preds = reg.predict(X_test)
y_preds[:10]
```

```python
acc_score = accuracy_score(y_test, y_preds)
print('The accuracy score of the logistic regression model is: {} '.format(acc_score))
```

```python
bal_acc_score = balanced_accuracy_score(y_test, y_preds)
print('The balance accuracy score of the logistic regression model is: {}'.format(bal_acc_score))
```

```python
print(confusion_matrix(y_test, y_preds))
```

```python
print(classification_report(y_test, y_preds))
```

```python
cm = confusion_matrix(y_test, y_preds)

fig = ConfusionMatrixDisplay(cm, display_labels = reg.classes_)

fig.plot()
```
